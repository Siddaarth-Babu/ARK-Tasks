```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

left_image = cv.imread('left.png')
right_image = cv.imread('right.png')

left_image_new = cv.cvtColor(left_image, cv.COLOR_BGR2GRAY)
right_image_new = cv.cvtColor(right_image, cv.COLOR_BGR2GRAY)
#left_image = cv.imread('items1_l.png', cv.IMREAD_GRAYSCALE)
#left_image = cv.imread('items1_r.png', cv.IMREAD_GRAYSCALE)

#left_image = cv.imread('items2_l.png', cv.IMREAD_GRAYSCALE)
#left_image = cv.imread('items2_r.png', cv.IMREAD_GRAYSCALE)

stereo = cv.StereoBM.create(numDisparities = 48, blockSize = 15)
#For each pixel algorithm will find the best disparity from 0
# Larger block size implies smoother, though less accurate disparity

depth = stereo.compute(left_image_new,right_image_new)

# cv.imshow("Left", left_image)
# cv.imshow("Right", right_image)

plt.imshow(depth)
plt.axis('off')
plt.show()
```
